```python
import matplotlib.pyplot as plt
import csv
import math
import numpy as np

rang = 5000
t = []
eng1 = []
eng2 = []
flow1 =[]
flow2 =[]
flow3 =[]
flow4 =[]
ftemp1 = []
ftemp2 = []
ftemp3 = []
ftemp4 = []
head = []
wind_spd = []
wind_ang = []
depth = []


  
with open('/content/drive/MyDrive/7902283_20211014_030000_minute_averages.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    next(lines)
    i = 0
    for row in lines:
        i+=1
        t.append(int(row[2][-5:-3])/60+int(row[2][-8:-6]))
        eng1.append(float(row[3]))
        eng2.append(float(row[4]))
        flow1.append(float(row[11]))
        flow2.append(float(row[12]))
        flow3.append(float(row[13]))
        flow4.append(float(row[14]))
        ftemp1.append(float(row[23]))
        ftemp2.append(float(row[24]))
        ftemp3.append(float(row[25]))
        ftemp4.append(float(row[26]))
        head.append(float(row[73]))
        wind_spd.append(float(row[74]))
        wind_ang.append(float(row[75]))
        depth.append(float(row[76]))
        #print(ftemp1)
        if (i == rang): break
```

```python
plt.plot(t, eng1, color = 'g', linestyle = 'dashed',
         marker = 'o',label = "Engine 1")
plt.plot(t, eng2, color = 'b', linestyle = '-',
         marker = '.',label = "Engine 2")
  
plt.xticks(rotation = 25)
plt.xlabel('Time (hour)')
plt.ylabel('Liter')
plt.title('Vessel Report: Engines', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(t, flow1, color = 'g', linestyle = 'dashed',
         marker = ".",label = "Engine 1 lit")
plt.plot(t, flow2, color = 'b', linestyle = '-',
         marker = ".",label = "Engine 2 lit")
plt.plot(t, flow3, color = 'r', linestyle = '-.',
         marker = ".",label = "Engine 2 lit")
plt.plot(t, flow4, color = 'y', linestyle = '--',
         marker = ".",label = "Engine 2 lit")
  
plt.xticks(rotation = 25)
plt.xlabel('Time (hour)')
plt.ylabel('Flow')
plt.title('Vessel Report: Flow', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(t, ftemp1, color = 'g', linestyle = 'dashed',
         marker = ".",label = "Engine 1 lit")
plt.plot(t, ftemp2, color = 'b', linestyle = '-',
         marker = ".",label = "Engine 2 lit")
plt.plot(t, ftemp3, color = 'r', linestyle = '-.',
         marker = ".",label = "Engine 2 lit")
plt.plot(t, ftemp4, color = 'y', linestyle = '--',
         marker = ".",label = "Engine 2 lit")
  
plt.xticks(rotation = 25)
plt.xlabel('Time (hour)')
plt.ylabel('temp')
plt.title('Vessel Report: Temp', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
## Question: Do we need to conside the heading or the angle is relative

wind_twr = []
wind_ort = []
for i in range(len(t)):
  wind_twr.append(math.sin((wind_ang[i]-head[i]*0)*math.pi/180)*wind_spd[i])
  wind_ort.append(math.cos((wind_ang[i]-head[i]*0)*math.pi/180)*wind_spd[i])


plt.plot(t, wind_twr, color = 'g', linestyle = 'dashed',
         marker = '*',label = "Toward wind")

plt.title('Vessel Report: Wind', fontsize = 20)
plt.xticks(rotation = 25)
plt.xlabel('Time (hour)')
plt.ylabel(' straight')
plt.figure()

plt.plot(t, wind_ort, color = 'b', linestyle = '-',
         marker = '*',label = "Ortogonal wind")
plt.title('Vessel Report: Wind ortogonal', fontsize = 20)  
plt.xticks(rotation = 25)
plt.xlabel('Time (hour)')
plt.ylabel('ortogonal')

plt.grid()
plt.legend()
plt.show()
```
